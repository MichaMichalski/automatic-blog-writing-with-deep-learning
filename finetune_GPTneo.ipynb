```python
from happytransformer import HappyGeneration
model_to_use = "EleutherAI/gpt-neo-125M"
# model_to_use = "EleutherAI/gpt-neo-1.3B"
happy_gen = HappyGeneration("GPT-NEO", model_to_use)
```

```python
from happytransformer import GENTrainArgs
```

```python
trainargs = GENTrainArgs(num_train_epochs=10)
```

```python
happy_gen.train('/content/drive/MyDrive/scrapes/scrapes_text.csv', args=trainargs)
```
